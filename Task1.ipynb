```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the Iris dataset
data = pd.read_csv('iris.csv')

# Creating a bar graph of 'sepal length (cm)'
# We'll create bins for the sepal length and then count the number of occurrences in each bin
bin_edges = list(range(int(data['sepal_length'].min()), int(data['sepal_length'].max()) + 1))
data['sepal_length'].value_counts(bins=bin_edges).sort_index().plot(kind='bar', figsize=(10,6), color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Sepal_Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length (cm)')

# Showing the plot
plt.grid(True)
plt.show()
```
